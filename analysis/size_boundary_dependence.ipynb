```julia
# Change this however you like.
# Examples:
# x_transform = x -> x
# x_transform = x -> x / 2048
# x_transform = x -> sqrt.(x)
# x_transform = x -> log10.(x)
w(u) = (1 - u)^0.001

panel = loaded_ev_sym_link.panels[1]  # size panel

fig = Figure(size = (900, 650))
ax = Axis(
    fig[1, 1];
    xscale = log10,
    yscale = log10,
    xlabel = L"f(j)",
    ylabel = L"\lambda_j",
    title = "ev_sym_link size panel",
)

#xlims!(ax, 1e-3, 1.5)
#ylims!(ax, 1e0, 1e2)

theme_colors_obs = CairoMakie.theme(:palette).color
theme_colors = theme_colors_obs isa Observables.Observable ? Observables.to_value(theme_colors_obs) : theme_colors_obs
colors = [theme_colors[mod1(i, length(theme_colors))] for i in 1:length(panel)]

for (i, s) in enumerate(panel)
    
    x_transform = x -> x/s.size
    y_transform = y -> y
    x = collect(2:length(s.mean))
    x = x_transform(x)

    mean = y_transform(s.mean[2:end])
    std = y_transform(s.std[2:end])
    ylo = y_transform(mean .- std)
    yhi = y_transform(mean .+ std)

    # Keep only valid points for log scales
    mask = (x .> 0) .& (mean .> 0)
    x = x[mask]
    mean =mean[mask]
    ylo =ylo[mask]
    yhi =yhi[mask]

    # Prevent lower/upper bands from dropping below 0 on log scale
    eps = minimum(mean) * 1e-3
    ylo = max.(ylo, eps)
    yhi = max.(yhi, eps)

    color = colors[i]
    label = "n=$(s.size)"

    band!(ax, x, ylo, yhi; color = (color, 0.18))
    lines!(ax, x, mean; color = color, linewidth = 3, label = label)
end

axislegend(ax; position = :rb)

fig
```
